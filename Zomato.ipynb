# Removing "NEW" , "-" and "/5" from Rate Column


```python
df1['rate'].unique()
```

```python
def function(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df1['rate'] = df['rate'].apply(function)
df1['rate'].head()
```

```python
df1['rate'].isnull()
```

# Removing (,) from Cost2Plates Column

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].apply(handlecomma)
df['approx_cost(for two people)'].unique()
```

# Cleaning Rest_types


```python
df1['rest_type'].value_counts()
```

```python
rest_types = df1['rest_type'].value_counts()
```

```python
rest_types_lessthan500 = rest_types[rest_types<500]
rest_types_lessthan500
```

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan500):
        return 'others'
    else:
        return value
    
df1['rest_type'] = df1['rest_type'].apply(handle_rest_type)
df1['rest_type'].value_counts()
```

# Cleaning Location Column

```python
df1.head()
```

```python
location = df1['location'].value_counts()

location_lessthan200 = location[location<200]



def handle_location(value):
    if(value in location_lessthan200):
        return 'others'
    else:
        return value
        
df1['location'] = df1['location'].apply(handle_location)
df1['location'].value_counts()
```

# Cleaning Cuisines Column

```python
cuisines = df1['cuisines'].value_counts()

cuisines_lessthan50 = cuisines[cuisines<50]



def handle_cuisines(value):
    if(value in cuisines_lessthan50):
        return 'others'
    else:
        return value
        
df1['cuisines'] = df1['cuisines'].apply(handle_cuisines)
df1['cuisines'].value_counts()
```
